```python
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import mplfinance as mplf
from mpl_finance import candlestick_ohlc
import datetime

df = yf.Tickers("BTU XLE DIA")
df2 = df.history(period = "1y")

df2["Close"].plot()
```

```python
df3 = yf.Ticker ("ZIM")
df4 = df3.history(period = "1y")
df5 = df4["Close"]

series_short = df5.rolling(window = 5, min_periods = 5).mean()
series_long = df5.rolling(window = 10, min_periods = 10).mean()

df_sma = pd.DataFrame(columns=['Short', 'Long'])
df_sma['Short'] = series_short
df_sma['Long'] = series_long
df_sma.plot(figsize = (12, 8))

```

```python
series_short2 = df5.ewm(span = 2).mean()
series_long2 = df5.ewm(span = 10).mean()

df_sma = pd.DataFrame(columns=['Short', 'Long'])
df_sma['Short'] = series_short2
df_sma['Long'] = series_long2
df_sma.plot(figsize = (12, 8))
```
